# Data Characteristics

## Prerequisites

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib.pyplot import *
```

```python
import matplotlib as mpl
mpl.rcParams['font.size'] = 15
```

### Loading the data

```python
hipAcc = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "hip left - Huawei", "acc.csv"), sep=";", decimal=",")
hipAcc = hipAcc.drop("Markiert:", axis=1)
hipAcc = hipAcc.drop("Unnamed: 5", axis=1)

hipGyro = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "hip left - Huawei", "gyro.csv"), sep=";", decimal=",")
hipGyro = hipGyro.drop("Markiert:", axis=1)
hipGyro = hipGyro.drop("Unnamed: 5", axis=1)

hipStep = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "hip left - Huawei", "step.csv"), sep=";", decimal=",")
hipStep = hipStep.drop("Markiert:", axis=1)
hipStep = hipStep.drop("Unnamed: 3", axis=1)

hip = pd.merge(hipAcc, hipGyro, on="Zeit in ms:")
hip = pd.merge(hip, hipStep, on="Zeit in ms:")
hip = hip.iloc[3:]
# trimming beginning to synchronize
hip['Zeit in ms:'] = hip['Zeit in ms:'].apply(lambda time: time - 60)
# inverting value because phone was upside down
hip['y in m/s²'] = hip['y in m/s²'].apply(lambda value: value * -1)
hip.head()
```

```python
leftFootAcc = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot left OnePlus", "acc.csv"), sep=";", decimal=",")
leftFootAcc = leftFootAcc.drop("Markiert:", axis=1)
leftFootAcc = leftFootAcc.drop("Unnamed: 5", axis=1)

leftFootGyro = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot left OnePlus", "gyro.csv"), sep=";", decimal=",")
leftFootGyro = leftFootGyro.drop("Markiert:", axis=1)
leftFootGyro = leftFootGyro.drop("Unnamed: 5", axis=1)

leftFootStep = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot left OnePlus", "step.csv"), sep=";", decimal=",")
leftFootStep = leftFootStep.drop("Markiert:", axis=1)
leftFootStep = leftFootStep.drop("Unnamed: 3", axis=1)

leftFoot = pd.merge(leftFootAcc, leftFootGyro, on="Zeit in ms:")
leftFoot = pd.merge(leftFoot, leftFootStep, on="Zeit in ms:")
leftFoot = leftFoot.iloc[2:]
# trimming beginning to synchronize
leftFoot['Zeit in ms:'] = leftFoot['Zeit in ms:'].apply(lambda time: time - 40)
leftFoot.head()
```

```python
rightFootAcc = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot right LG", "acc.csv"), sep=";", decimal=",")
rightFootAcc = rightFootAcc.drop("Markiert:", axis=1)
rightFootAcc = rightFootAcc.drop("Unnamed: 5", axis=1)

rightFootGyro = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot right LG", "gyro.csv"), sep=";", decimal=",")
rightFootGyro = rightFootGyro.drop("Markiert:", axis=1)
rightFootGyro = rightFootGyro.drop("Unnamed: 5", axis=1)

rightFootStep = pd.read_csv(os.path.join("..", "data", "17-03-19", "measurement-3", "foot right LG", "step.csv"), sep=";", decimal=",")
rightFootStep = rightFootStep.drop("Markiert:", axis=1)
rightFootStep = rightFootStep.drop("Unnamed: 3", axis=1)

rightFoot = pd.merge(rightFootAcc, rightFootGyro, on="Zeit in ms:")
rightFoot = pd.merge(rightFoot, rightFootStep, on="Zeit in ms:")
rightFoot.head()
```

## Combined Gait Analysis

### Merge data from measurement points

```python
combinedDf = hip.merge(leftFoot, how='outer', on="Zeit in ms:", suffixes=('_hip', '_leftFoot'))
combinedDf = combinedDf.merge(rightFoot, how='outer', on="Zeit in ms:", suffixes=('', '_rightFoot'))
```

### Jumping for synchronization

```python
fig, ax = subplots()
filteredCombinedDf2 = combinedDf.loc[(combinedDf['Zeit in ms:'] < 25000) & (combinedDf['Zeit in ms:'] > 18000)]
jumpingPlot = filteredCombinedDf2.set_index("Zeit in ms:")[['y in m/s²_hip', 'z in m/s²_leftFoot', 'z in m/s²']].plot(figsize=(20, 10), ax=ax)
jumpingPlot.set(xlabel='Time in ms', ylabel='Vertical movement in m/s²', title='Measured movement during jumping for synchronization')
ax.legend(['Vertical Acceleration Hip', 'Vertical Acceleration Left Foot', 'Vertical Acceleration Right Foot'])
```

```python
fig, ax = subplots()
filteredCombinedDf = combinedDf
completeGaitPlot = filteredCombinedDf.set_index("Zeit in ms:")[['y in m/s²_hip', 'z in m/s²_leftFoot', 'z in m/s²', 'y in rad/s_hip']].plot(figsize=(20, 10), ax=ax)
completeGaitPlot.set(xlabel='Time in ms', ylabel='Vertical movement', title='Measured movement during the whole measurement')
ax.legend(['Vertical Acceleration Hip in m/s²', 'Vertical Acceleration Left Foot in m/s²', 'Vertical Acceleration Right Foot in m/s²', 'Vertical Rotation Hip in rad/s'])
completeGaitPlot.get_figure().savefig(os.getcwd()+'/plots/completeGaitPlot.svg', bbox_inches="tight", format="svg", dpi=300)
```

```python
fig, ax = subplots()
normalGaitDf = combinedDf.loc[(combinedDf['Zeit in ms:'] < 44000) & (combinedDf['Zeit in ms:'] > 33500)]
normalGaitPlot = normalGaitDf.set_index("Zeit in ms:")[['y in m/s²_hip', 'z in m/s²_leftFoot', 'z in m/s²', 'y in rad/s_hip']].plot(figsize=(20, 10), ax=ax)
normalGaitPlot.set(xlabel='Time in ms', ylabel='Vertical movement', title='Measured movement during the normal gait measurement')
ax.legend(['Vertical Acceleration Hip in m/s²', 'Vertical Acceleration Left Foot in m/s²', 'Vertical Acceleration Right Foot in m/s²', 'Vertical Rotation Hip in rad/s'])
normalGaitPlot.get_figure().savefig(os.getcwd()+'/plots/normalGaitPlot.svg', bbox_inches="tight", format="svg", dpi=300)
```

```python
fig, ax = subplots()
weirdWaistGaitDf = combinedDf.loc[(combinedDf['Zeit in ms:'] < 85000) & (combinedDf['Zeit in ms:'] > 67000)]
weirdWaistPlot = weirdWaistGaitDf.set_index("Zeit in ms:")[['y in m/s²_hip', 'z in m/s²_leftFoot', 'z in m/s²', 'y in rad/s_hip']].plot(figsize=(20, 10), ax=ax)
weirdWaistPlot.set(xlabel='Time in ms', ylabel='Vertical movement', title='Measured movement during pelvic obliquity measurement')
ax.legend(['Vertical Acceleration Hip in m/s²', 'Vertical Acceleration Left Foot in m/s²', 'Vertical Acceleration Right Foot in m/s²', 'Vertical Rotation Hip in rad/s'])
weirdWaistPlot.get_figure().savefig(os.getcwd()+'/plots/weirdWaistPlot.svg', bbox_inches="tight", format="svg", dpi=300)
```

```python
fig, ax = subplots()
limpingGaitDf = combinedDf.loc[(combinedDf['Zeit in ms:'] < 116000) & (combinedDf['Zeit in ms:'] > 95000)]
limpingPlot = limpingGaitDf.set_index("Zeit in ms:")[['y in m/s²_hip', 'z in m/s²_leftFoot', 'z in m/s²', 'y in rad/s_hip']].plot(figsize=(20, 10), ax=ax)
limpingPlot.set(xlabel='Time in ms', ylabel='Vertical movement', title='Measured movement during limping measurement')
ax.legend(['Vertical Acceleration Hip in m/s²', 'Vertical Acceleration Left Foot in m/s²', 'Vertical Acceleration Right Foot in m/s²', 'Vertical Rotation Hip in rad/s'])
limpingPlot.get_figure().savefig(os.getcwd()+'/plots/limpingPlot.svg', bbox_inches="tight", format="svg", dpi=300)
```
